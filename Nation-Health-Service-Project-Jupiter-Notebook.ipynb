Objective 2;
Create four visualisations indicating the number of appointments for service setting per season. The seasons are summer (June to August 2021), autumn (September to November 2021), winter (December to February 2022), and spring (March to May 2022).

```python
# Define a function to categorize months into seasons
def get_season(month):
    if month in [6, 7, 8]:
        return 'Summer 2021'
    elif month in [9, 10, 11]:
        return 'Autumn 2021'
    elif month in [12, 1, 2]:
        return 'Winter 2021-2022'
    elif month in [3, 4, 5]:
        return 'Spring 2022'

# Apply the function to create a new 'season' column
nc_ss['season'] = nc_ss['month'].apply(get_season)
```

```python
#create new dataframe that has the added 'seasons' column
nc_ss_day = nc_ss.groupby(['season', 'service_setting'])['count_of_appointments'].sum().reset_index()
nc_ss_day
```

```python
#1. We can use a grouped barcahrt to visualize the above information
plt.figure(figsize=(12, 8))
sns.barplot(x='season', y='count_of_appointments', hue='service_setting', data=nc_ss_day)
plt.title('Number of Appointments per Service Setting by Season')
plt.ylabel('Count of Appointments (Millions)')
plt.xlabel('Season')
plt.xticks(rotation=45)
plt.legend(title='Service Setting', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('number_of_appointments_per_service_setting_by_season_bar_plot.png', format='png', dpi=300)

plt.show()
```

The grouped boxplot above shows the General practice setting as having the highest number of appointments across alll seasons and the Extended Aceess Provision service setting having the lowest.

```python
#We can also plot a stacked barchart to visualise the data.
# First, we create a pivot table for stacked bar chart
pivot_nc_ss_day = nc_ss_day.pivot_table(index='season', columns='service_setting', values='count_of_appointments')

# Plotting the stacked bar chart
pivot_nc_ss_day.plot(kind='bar', stacked=True, figsize=(15, 12))
plt.title('Number of Appointments per Service Setting by Season')
plt.ylabel('Count of Appointments(Millions')
plt.xlabel('Season')
plt.xticks(rotation=45)
plt.legend(title='Service Setting', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('number_of_appointments_per_service_setting_by_season_stacked_bar.png', format='png', dpi=300)

plt.show()
```

Just like the grouped bar chart, the general practice is leading in the highest number of appointments per season while Extended Access Provision has the least. 

```python
#3. We can also use line plot to show the trend in appointment numbers across the seasons
plt.figure(figsize=(15, 12))
sns.lineplot(x='season', y='count_of_appointments', hue='service_setting', data=nc_ss_day, marker='o')
plt.title('Trend of Appointments per Service Setting by Season')
plt.ylabel('Count of Appointments (Millions)')
plt.xlabel('Season')
plt.xticks(rotation=45)
plt.legend(title='Service Setting', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig('trend_of_appointments_per_service_setting_by_season_line_plot.png', format='png', dpi=300)

plt.show()
```

Across all service settings, there was a dip in appointment numbers in summer 2021. Autumn saw the highest number of appointments across all service sectors.

```python
#4. Finally, a clusterd heatmap to visualize the data
sns.clustermap(nc_ss_day.pivot_table(index='season', columns='service_setting', values='count_of_appointments'), cmap='coolwarm')
plt.title('Appointments by Service Setting Across Seasons')
plt.savefig('appointments_by_service_setting_across_seasons.png', format='png', dpi=300)

plt.show()
```

The heatmap above shows the secotors and seasons with high number of appointements in red and low ones in blue.

```python
# Group by the selected columns and calculate the sum of 'count_of_appointments'
ar_agg = ar.groupby(['appointment_month', 'hcp_type', 'appointment_status', 'appointment_mode', 'time_between_book_and_appointment'])['count_of_appointments'].sum().reset_index()

# View the new DataFrame
ar_agg.head()
```

```python
#Determine average utilisation of services:
```

```python
# Group by 'appointment_month' and sum 'count_of_appointments'
ar_df = ar_agg.groupby('appointment_month')['count_of_appointments'].sum().reset_index()
```

```python
# Calculate the utilisation as a percentage of the maximum capacity
ar_df['utilisation'] = (ar_df['count_of_appointments'] / 30) / 1200000 * 100
```

```python
# Round the utilisation to one decimal place
ar_df['utilisation'] = ar_df['utilisation'].round(1)

# View the new DataFrame
ar_df
```

```python
#Create two lineplots with Seaborn:
```

```python
# Convert 'appointment_month' to string
ar_df['appointment_month'] = ar_df['appointment_month'].astype(str)
# Set the style of seaborn plots
sns.set(style='whitegrid')
```

```python
# Create the first lineplot for number of monthly visits
plt.figure(figsize=(15, 12))
sns.lineplot(x='appointment_month', y='count_of_appointments', data=ar_df, marker='o')
plt.title('Number of Monthly Visits')
plt.xlabel('Month')
plt.ylabel('Number of Appointments (Millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create the second lineplot for monthly capacity utilisation
plt.figure(figsize=(15, 12))
sns.lineplot(x='appointment_month', y='utilisation', data=ar_df, marker='o', color='orange')
plt.title('Monthly Capacity Utilisation')
plt.xlabel('Month')
plt.ylabel('Utilisation (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_capacity_utilisation.png', format='png', dpi=300)

plt.show()
```

```python
# Define the function to convert time intervals to numeric days
def convert_to_days(time_str):
    try:
        if 'Day' in time_str:
            return int(time_str.split()[0])
        elif 'to' in time_str:
            parts = time_str.split('to')
            return (int(parts[0].strip().split()[0]) + int(parts[1].strip().split()[0])) / 2
        else:
            # Handle unexpected values
            return None
    except ValueError:
        return None
```

```python
# Apply the conversion function and handle missing values
ar_agg['time_between_book_and_appointment_days'] = ar_agg['time_between_book_and_appointment'].apply(convert_to_days)
```
